```python
import pandas as pd
df=pd.read_excel('monthlyannualprecip.xlsx',header=1)
df.head(n=10)
```

```python
df_new=df.copy()
df_new['Decade'] = (10 * (df_new['YEAR'] // 10)).astype(str) + 's'
df_new.head(13)
```

```python
df_new2=df_new.groupby('Decade')['ANNUAL'].mean().to_frame(name = 'Average Precip').reset_index()
df_new2
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
plt.bar(x=df_new2['Decade'],height=df_new2['Average Precip'])
plt.plot(df_new2['Decade'],df_new2['Average Precip'],marker='*',ms=10)
```
