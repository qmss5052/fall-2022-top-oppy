```python
import pandas as pd

data = pd.read_csv('/Users/aileen/Desktop/TOP/svi_score.csv')



data.info()
```

```python
from matplotlib import pyplot as plt

area = data[data["area/ny"]=="area"]
ny = data[data["area/ny"]=="ny"]

x = [2000,2010,2014,2016,2018,2020]
plt.plot("year", "SVI score", data = area, label = "Census Tract for Hernandez/Hughes School", color = "green")
plt.xticks(x)
plt.plot("year", "SVI score", data = ny, label = "New York county", color = "blue")
plt.legend(loc = 'best')
plt.title("Social Social Vulnerability Index among Census Tract \n for Hernandez/Hughes School vs New York county")
plt.xlabel("year")
plt.ylabel("SVI score %")
plt.show()
```
